```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(pd.read_csv('enjoysport.csv'))
concept = np.array(data.iloc[:,0:-1])
target = np.array(data.iloc[:,-1])
```

```python
s = concept[0].copy()
g = [["?" for _ in range(len(s))] for g in range(len(s))]

for i,h in enumerate(concept):
    if target[i] == 'yes':
        for x in range(len(s)):
            if h[x] != s[x]:
                s[x]='?'
                g[x][x]='?'
    
    else:
        for x in range(len(s)):
            if h[x]!=s[x]:
                g[x][x]=s[x]
            else:
                g[x][x]='?'
    
    print(f"Step {i+1}: ")
    print(f"S: {s}")
    print(f"G: {g}")

g = [h for h in g if set(h) != {'?'}]
print(f"Final S: {s}")
print(f"Final G: {g}")
```
