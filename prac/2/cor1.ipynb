```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species'] = iris.target
```

```python
numeric_df = df.drop(columns=['Species'], axis=1)
corr_m = numeric_df.corr()
corr_m.round(2)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_m,annot=True,fmt='.2f',cmap='coolwarm',linewidths=0.5)
plt.title('Heatmap',fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
strong_pairs = corr_m.unstack().sort_values(ascending=False)
strong_pairs = strong_pairs[(abs(strong_pairs)>0.7) & (abs(strong_pairs)<1.0)]
```

```python
seen = set()
for (f1,f2),val in strong_pairs.items():
    if(f2,f1) not in seen:
        print(f"{f1} - {f2} : {val:.2f}")
        seen.add((f1,f2))
```
