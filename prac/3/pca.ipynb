```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
print(pca.explained_variance_ratio_)
print(f"Total explained variance: {(sum(pca.explained_variance_ratio_)*100):.2f}%")
```

```python
plt.figure(figsize=(10,8))
labels=iris.target_names
colors=['navy','red', 'darkorange']

for i in[0,1,2]:
    plt.scatter(X_pca[iris.target==i,0],
                X_pca[iris.target==i,1],
                label=labels[i],
                color=colors[i],
                alpha=0.8
                )
plt.xlabel(f"PC1: {pca.explained_variance_ratio_[0]*100:.2f}%")
plt.ylabel(f"PC2: {pca.explained_variance_ratio_[1]*100:.2f}%")
plt.figtext(0.02,0.02,"PC1 influenced by Petal length \nPC2 influenced by Sepal width")
plt.legend()
plt.show()
```

```python
loadings = pd.DataFrame(pca.components_.T, 
                        columns=['PC1', 'PC2'], 
                        index=iris.feature_names)
print(loadings)
```
