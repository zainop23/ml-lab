```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('EnjoySport.csv')
X = data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
S=X[0].copy()

G=[["?" for _ in S] for _ in S]
```

```python
for i in range(len(X)):
    sample = X[i]

    if y[i]=='Yes':
        for j in range(len(S)):
            if sample[j]!=S[j]:
                S[j]='?'
                G[j] = ['?'] * len(S)
            
    else :
        for j in range(len(S)):
            if sample[j]!=S[j]:
                G[j][j] = S[j]
            else :
                G[j][j]='?'
        
    print(f"Step {i+1}: ")
    print("S:", S)
    print("G:", G)
```

```python
G=[g for g in G if g != ['?']*len(S)]

print("S:", S)
print("G:", G)
```
