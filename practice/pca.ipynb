```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
scaler=StandardScaler()
X_scaled = scaler.fit_transform(iris.data)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
print(pca.explained_variance_ratio_)
print("total explained variance ratio {:.2f}%".format(sum(pca.explained_variance_ratio_)*100))
```

```python
plt.figure(figsize=(15,10))
colors=['navy', 'red', 'darkorange']
labels = iris.target_names

for i in [0,1,2]:
    plt.scatter(
        X_pca[iris.target==i,0],
        X_pca[iris.target==i,1],
        label=labels[i],
        color=colors[i],
        alpha=0.8
    )

plt.title('PCA on Iris Dataset')
plt.xlabel(f"PC1 {pca.explained_variance_ratio_[0]*100 :.2f}%")
plt.ylabel(f"PC2 {pca.explained_variance_ratio_[1]*100 :.2f}%")
pc1_top='petal length (cm)'
pc2_top='sepal width (cm)'
plt.figtext(0.02,0.02, f"PC1: {pc1_top}\nPc2: {pc2_top}",
            bbox = dict(facecolor='white'))
plt.legend()
plt.grid()
plt.show()
```
