```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame(pd.read_csv("enjoysport.csv"))
concept = np.array(data.iloc[:,0:-1])
target = np.array(data.iloc[:,-1])
print(concept)
```

```python
s=concept[0].copy()
g=[["?" for _ in range(len(s))] for g in range(len(s))]
print("inital s and g")
print(s)
print(g)
for i,h in enumerate(concept):
    if target[i]=='yes':
        for x in range(len(s)):
            if h[x]!=s[x]:
                s[x]='?'
                g[x][x]='?'
    else:
        for x in range(len(s)):
            if h[x]!=s[x]:
                g[x][x]=s[x]
            else:
                g[x][x]="?"
    print("step ",(i+1))
    print(s)
    print(g)

indices=[i for i,val in enumerate(g) if val==['?','?','?','?','?','?']]
for i in sorted(indices,reverse=True):
    del g[i]
print("final")
print(s)
print(g)
```
